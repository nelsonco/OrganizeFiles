```python
import os
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Parent Directory
parent_dir = "C:/Users/Court/Downloads/PDFDocuments"
dir_list = os.listdir(parent_dir)
```

```python
class FileInfo:
    def __init__(self,file):
        self.file = file
        file_terms = file.replace('_',' ')
        file_terms = file_terms.replace('-',' ')
        file_terms = file_terms.split('.')
        file_terms = file_terms[:-1]
        self.file_terms = file_terms[0]
        
file_info=[]
for file in dir_list:
    file_info.append(FileInfo(file).file_terms)
```

```python
print(file_info[10])
```

```python
data = pd.DataFrame(file_info)
```

```python
data
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(file_info)
```

```python
X = X.toarray()
```

```python
print(vectorizer.get_feature_names())
```

```python
df = pd.DataFrame(data=X,columns = vectorizer.get_feature_names())
```

```python
df
```

```python
sumdf = pd.DataFrame(X.sum(axis=0).reshape(1,len(X[0])),columns = vectorizer.get_feature_names())
```

```python
sumdf
```

```python
df2 = pd.concat([sumdf,df]).reset_index(drop=True)
```
