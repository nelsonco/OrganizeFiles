```python
import os
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import enchant
```

```python
# Parent Directory
parent_dir = "C:/Users/Court/Downloads/PDFDocuments"
dir_list = os.listdir(parent_dir)
```

```python
class FileInfo:
    def __init__(self,file):
        self.file = file
        file_terms = file.replace('_',' ')
        file_terms = file_terms.replace('-',' ')
        file_terms = file_terms.split('.')
        file_terms = file_terms[:-1]
        self.file_terms = file_terms[0]

file_info=[]
files_list=[]
for file in dir_list:
    if os.path.isfile(parent_dir+"/"+file):
        file_info.append(FileInfo(file).file_terms)
        files_list.append(file)
```

```python
stop_words = ["courtney","nelson","and"]
```

```python
vectorizer = CountVectorizer(stop_words=stop_words)
```

```python
X = vectorizer.fit_transform(file_info)
```

```python
X = X.toarray()
```

```python
print(vectorizer.get_feature_names())
```

```python
df = pd.DataFrame(data=X,columns = vectorizer.get_feature_names())
```
